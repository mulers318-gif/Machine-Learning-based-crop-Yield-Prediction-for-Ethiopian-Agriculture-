### 3.6 Outlier Treatment

Since your dataset contains suspiciously high yield values, investigate them first.

```python
df_clean.sort_values(
    "Yeild_kg/ha",
    ascending=False
).head()
```

##### Optional:Extreme Outlier Using IQR

```python
q1=df_clean["Yeild_kg/ha"].quantile(0.5)
q3=df_clean["Yeild_kg/ha"].quantile(0.75)
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
upper=q3+1.5*IQR
```

```python
# Use the '&' operator and match the exact column name with a space
df_clean = df_clean[
    (df_clean["Yeild_kg/ha"] >= lower) & 
    (df_clean["Yeild_kg/ha"] <= upper)
]

print("Outliers filtered successfully! Current data shape:", df_clean.shape)
```

The Interquartile Range (IQR) method was used to identify and remove extreme outliers.

# PART 5:Machine Learning Modeling

This section builds predictive models to estimate crop yield.

The target variable is:

Yeild_kg/ha

### 5.1 Objectives

##### Machine Learning

The objective of this phase is to develop machine learing models capable of predicting crop yield.

Three models will be evaluated:

1.Linear Regression

2.Decision Tree regressor

3.Random Forest Regressor

Their performance will be compared using:

-MAE(Mean Absolute Error)

-RMSE(Root Mean Squared error)

-R$^2$ Score

### 5.2 Import Machine Learning Libraries

```python
# Machine Learning Models
from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor

# Model evaluation
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
# Numerical Operations
import numpy as np
```

### 5.3 Verfiy Training and Testing Data 

```python
# Display dimensions of training and testing data
print("Training Features:",x_train.shape)
print("Testing Features:",x_test.shape)
print("Training Target:",y_train.shape)
print("Test Target:",y_test.shape)
```

The dataset has already been divided into:

-Training data(80%)

-Testing data (20%)

The training set is used for learning patterns.

The testing set is used to evaluate performance on unseen data.

### Model 1:Linaer Regression

```python
# Creat Linear regression Model
lr=LinearRegression()
# Train the model
lr.fit(x_train,y_train)
```

```python
# Generate Predictions
pred_lr=lr.predict(x_test)
```

```python
# Evaluate Linear Regression
mae_lr=mean_absolute_error(
    y_test,
    pred_lr
)
rmse_lr=np.sqrt(
    mean_squared_error(
        y_test,
        pred_lr
    )
)
r2_lr=r2_score(
    y_test,
    pred_lr
)
print("Linear Regression Results")
print("MAE=",mae_lr)
print("RMSE=",rmse_lr)
print("R2=",r2_lr)
```

Linear Regression assumes a linear relationship between input variables and crop yield.

It serves as a baseline model for comparison

### Model 2:Decission Tree Regressor

```python
# Create Decissin Tree model
dt=DecisionTreeRegressor(
    random_state=42
)
# Train the model
dt.fit(x_train,y_train)
```

```python
# Make predictions
pred_dt=dt.predict(x_test)
```

```python
# Evaluate Decision Tree
mae_dt=mean_absolute_error(
    y_test,
    pred_dt
)
rmse_dt=np.sqrt(
    mean_squared_error(
        y_test,
        pred_dt
    )
)
r2_dt=r2_score(
    y_test,
    pred_dt
)
print("Decission Tree Results")
print("MAE=",mae_dt)
print("RMSE=",rmse_dt)
print("R2=",r2_dt)
```

Decision Trees can capture nonlinear relationships and interactions between variables.

However,they may overfit the training data.

### Model 3:Random Forest Regressor

```python
# Create Random Forest model
rf=RandomForestRegressor(
    n_estimators=100,
    random_state=42
)
# Train model
rf.fit(
    x_train,
    y_train
)
```

```python
# Predict Crop Yield
pred_rf=rf.predict(x_test)
```

```python
# Evaluate Random Forest
mae_rf=mean_absolute_error(
    y_test,
    pred_rf
)
rmse_rf=np.sqrt(
    mean_squared_error(
        y_test,
        pred_rf
    )
)
r2_rf=r2_score(
    y_test,
    pred_rf
)
print("Random Forest Results")
print("MAE=",mae_rf)
print("RMSE=",rmse_rf)
print("R2=",r2_rf)
```

Random Forest combines multiple decission trees to improve prediction accuracy and reduce overfitting.

It is widely used in agricultural prediction problems.

### 5.4 Compare Model Performance

```python
# Create comparison table
results=pd.DataFrame({
    "Model":[
        "Linear Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "MAE":[
        mae_lr,
        mae_dt,
        mae_rf
    ],
    "RMSE":[
        rmse_lr,
        rmse_dt,
        rmse_rf
    ],
    "R2":[
        r2_lr,
        r2_dt,
        r2_rf
    ]
})
results
```

```python
# Sort models by R^2 score
results.sort_values(
    by="R2",
    ascending=False
)
```

The best model is the one with:

-Lowest MAE

-Lowest RMSE

-Highest R$^2$
